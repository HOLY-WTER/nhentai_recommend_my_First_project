```python
def BF(doujin_id,sort_id):
    low=0
    mid = 0
    high=len(sort_id)-1
    #print('{0}={1} is {2}'.format(type(doujin_id),"doujin_id",doujin_id))
    doujin_id=int(doujin_id)
    #print(sort_id)
    while low<=high:
            mid = (high + low) // 2
            #print(type(sort_id[mid]))
           # print(sort_id[mid])
            if sort_id[mid]>doujin_id:
                high=mid-1
            elif sort_id[mid]<doujin_id:
                low=mid+1
            else:
                #print('{0} is {1}'.format("mid",sort_id[mid]))
                return mid
    return -1
    
```
